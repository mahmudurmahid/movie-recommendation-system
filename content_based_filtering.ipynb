# Content-Based Filtering

### Load the Data

```
import pandas

movies = pandas.read_csv("movies.csv")
```

```
movies
```

### Term Frequency & Inverse Document Frequency

```
import sklearn # also known as scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
tfidf = TfidfVectorizer(stop_words="english")

movies["overview"] # access overview column in movies.csv
```

```
movies["overview"] = movies["overview"].fillna("") # fill null values with empty string

tfidf_matrix = tfidf.fit_transform(movies["overview"]) # create tf-idf matrix to show similarity of words used in overview of movies
pandas.DataFrame(tfidf_matrix.toarray(), columns = tfidf.get_feature_names()) # convert & display tf-idf matrix
```

### Similarity Matrix

```
from sklearn.metrics.pairwise import linear_kernel

similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
similarity_matrix # shows similarity of words used in overview of movies
```

### Extracting Most Similar Movies

```
def similar_movies(movie_title, nr_movies):
    idx = movies[movies["title"] == movie_title].index[0] # get index of movie
    scores = list(enumerate(similarity_matrix[idx])) # get similarity scores
    scores = sorted(scores, key=lambda x: x[1], reverse=True) # sort scores in descending order
    movie_indices = [tpl[0] for tpl in scores[1:nr_movies+1]] # get indices of most similar movies
    similar_movie_title = movies["title"].iloc[movie_indices] # get titles of most similar movies
    return similar_movie_title
```

```
similar_movies("Spider-Man", 10)
```
